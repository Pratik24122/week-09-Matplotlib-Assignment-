Q1

Matplotlib is an amazing visualization library in Python for 2D plots of arrays.

Line Plot: Connects data points using lines.

Bar Plot: Displays bars to represent data values.

Scatter Plot: Shows individual data points without connecting lines.

Pie Chart: Represents data as slices of a circular pie.

Area Plot: Fills the area under the curve of a line plot.

Q2

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

Scatter Plot: Shows individual data points without connecting lines.

```python
import matplotlib.pyplot as plt
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.scatter(x,y)
plt.show()
```

Q3

The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure. It allows you to arrange multiple plots in a grid-like structure, making it easier to compare and visualize different data.

```python
#line1
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('Line 1')

#line2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title('Line 2')

#line3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title('Line 3')

#line4
x = np.array([0, 1, 2, 3, 4, 5])
y= np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title('Line 4')

plt.tight_layout()

plt.show()
```
